```python
import pandas as pd
import bnlearn as bn
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("preprocessing/data_2.csv")
```

```python
A = bn.structure_learning.fit(df=data,methodtype='hc',root_node='class',scoretype='k2')
```

```python
A['adjmat']
```

```python
bn.plot(model=A,node_size=400,title='Network (A)')
plt.show()
```
