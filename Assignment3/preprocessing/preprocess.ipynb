```python
import pandas as pd

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("../wine/wine.data")
```

```python
X = df.drop(columns='class')
Y = df['class']
```

```python
X.describe()
```

```python
Y.describe()
```

```python
from sklearn.preprocessing import KBinsDiscretizer

col_n_bins = {
    'alcohol':3,
    'malicacid':6,
    'ash':5,
    'alcalinity':10,
    'Mg':12,
    'phenols':6,
    'flav':10,
    'nonflav':6,
    'proantho':7,
    'clr_intensity':12,
    'hue':6,
    '0D280_0D315':6,
    'proline':15
}

X_disc = pd.DataFrame()
for col,n_bins in col_n_bins.items():
    discretizer = KBinsDiscretizer(n_bins=n_bins,encode='ordinal',strategy='uniform')
    X_disc[col] = discretizer.fit_transform(X[[col]]).flatten()

X_disc
```

```python
df_disc = pd.concat([X_disc,Y],axis=1)
df_disc.to_csv("data.csv",index=False)
```
